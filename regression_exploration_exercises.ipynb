3. Write a function named plot_variable_pairs that accepts a dataframe as input and plots all of the pairwise relationships along with the regression line for each pair.

```python
count_var = ['bedrooms', 'bathrooms', 'square_footage', 'tax_value', 'year_built', 'tax_amount', 'fips']
disc_var = train['fips']
cont_var = ['bedrooms', 'bathrooms', 'square_footage', 'tax_value', 'year_built', 'tax_amount']
```

```python
def plot_variable_pairs(train):
    '''
    This function takes in a df as input and plots 
    the relationships between variables
    with a regression line for each pair
    '''
    sns.pairplot(data=train.sample(100), kind='reg', corner=True)
    plt.title('distribution')
    plt.show()
        
plot_variable_pairs(train)
```

4. Write a function named plot_categorical_and_continuous_vars that accepts your dataframe and the name of the columns that hold the continuous and categorical features and outputs 3 different plots for visualizing a categorical variable and a continuous variable.

```python
def plot_categorical_and_continuous_vars(train):
    '''
    This function takes in a df and all columns as input and
    returns three different plots for visualizing categorical
    variables and continuous variables
    '''
    for col in cont_var:
    
        sns.violinplot(x=disc_var, y=train[col], data=train.sample(100))
        plt.title('distribution')
        plt.show()
        
        sns.boxplot(x=disc_var, y=train[col], data=train.sample(100))
        plt.title('distribution')
        plt.show()
        
        sns.barplot(x=disc_var, y=train[col], data=train.sample(100))
        plt.title('distribution')
        plt.show()
        
plot_categorical_and_continuous_vars(train)
```
